```r
library("readxl")
library("UsingR")
```

```r
covid<-read_xlsx("Daily Update Data Agregat Covid-19 Jakarta (2).xlsx")
```

```r
head(covid)
covid<-data.frame(covid)
```

```r
mean(covid[,12])
median(covid[,12])
```

```r
getmode <- function(v) {
uniqv <- unique(v)
uniqv[which.max(tabulate(match(v, uniqv)))]
}
result <- getmode(covid[,12])
print(result)
```

```r
min(covid[,12])
max(covid[,12])
```

```r
boxplot(covid[,12])
```

```r
outlier = boxplot.stats(covid[,12])$out
outlier
```

```r
plot(covid[,12],covid[,13], xlab = "Positif harian",
     ylab = "Positif aktif", col=2, pch=16, main='')
abline(lm(covid[,13] ~ covid[,12]),  col='blue')
```

```r
reg1 <- lm(covid[,13] ~ covid[,12])
reg1
```

```r
summary(reg1)
```

```r
cor(covid[, 12], covid[, 13])
```
