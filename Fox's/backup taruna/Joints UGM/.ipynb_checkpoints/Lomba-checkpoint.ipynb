```python
kat_umur=[]
for umur in train["age"]:
    if umur>=17 and umur<27:
        kat_umur.append(1)
    if umur>=27 and umur<44:
        kat_umur.append(2)
    if umur>=44 and umur<55:
        kat_umur.append(3)
    if umur>=55:
        kat_umur.append(4)
train["kat_umur"]=kat_umur
```

```python
import scipy
from scipy.stats import chi2
x=train.kat_umur
y=train["Best Performance"]
contingency_table=pd.crosstab(y,x)
plt.figure(figsize=(14,10))
sns.heatmap(contingency_table, annot=True, annot_kws={"size":30})
plt.xlabel('Umur', fontsize = 35) # x-axis label with fontsize 15
plt.ylabel('Best Performance', fontsize = 35) # y-axis label with fontsize 15
plt.xticks(fontsize=30,rotation=10)
plt.yticks(fontsize=30,rotation=0)
plt.title("Heatmap Kontingensi Job Level vs Person Level",fontsize=40)
plt.show()

print()
print("Tes Independensi")
print()
print("H0 : Dua Variabel saling independen")
print("H1 : Dua Variabel saling berhubungan ")
print()
alpha=0.05
no_of_rows=len(contingency_table.iloc[:,0])
no_of_columns=len(contingency_table.iloc[0,:])
deg=(no_of_rows-1)*(no_of_columns-1)
Observed_Values = contingency_table.values 
b=scipy.stats.chi2_contingency(contingency_table)
Expected_Values = b[3]
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
critical_value=chi2.ppf(q=1-alpha,df=deg)
p_value=1-chi2.cdf(x=chi_square_statistic,df=deg)
print('Level Signifikansi: ',alpha)
print('Derajat Kebebasan: ',deg)
print('chi-square statistic:',chi_square_statistic)
print('critical_value:',critical_value)
print('p-value:',p_value)
print()
print("Kesimpulan:")
if chi_square_statistic>=critical_value:
    print("Tolak H0, Dua Variabel saling berhubungan")
else:
    print("Terima H0, Dua Variabel saling independen")
```

```python
train_data = pd.concat([train_data, pd.get_dummies(train_data["Jenis JG Job Level"]),pd.get_dummies(train_data["Jenis JG Person Level"]), pd.get_dummies(train_data["Jenis JG Branch Level"]) ] , axis = 1)
```

```python
train_data.dtypes
```

```python
train_data=train_data.dropna()
```

```python
def training_data(fitur):
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import mean_squared_error
    from sklearn import metrics
    X=train_data[fitur]
    y=train_data["Best Performance"]
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
    
    #XGBClassifier
    from xgboost import XGBRegressor
    modelXGB = XGBRegressor(random_state = 1)
    modelXGB.fit(train_X, train_y)
    error = mean_squared_error(modelXGB.predict(val_y), val_y)

    return error
```

```python
def tulis(fitur):
    skor=training_data(fitur)
    print("Error XGBRegressor : ",skor)
```
