```python
def ambil_data_Single (kolom):
    hasil =[0 for i in range(2021)]
    for i in list(kolom):
        temp = data[i]
        for j in range(2021):
            if (temp[j] !=0):
                if hasil[j]==0:
                    hasil[j] = temp[0]
    for i in range(2021):
        if hasil[i]==0:
            hasil[i]='Tidak Mengisi'
    return (hasil)
```

```python
def ambil_data_Multi(kolom,kata):
    hasil = [[] for i in range (2021)]
    for i in list(kolom):
        temp = data[i]
        for i in range (2021):
            if (temp[i] !=0):
                hasil[i] = hasil[i] +[kata + str(temp[0])]
            else :
                hasil[i] = hasil[i] +['Tidak '+kata+ str(temp[0])]
    return hasil
```
