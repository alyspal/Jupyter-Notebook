```python
import pandas  as pd
df = pd.read_csv("athlete_events.csv")

(df)
```

```python
#Deskripsi Data dan File

#Deskripsi

'''
The file athlete_events.csv contains 271116 rows and 15 columns. Each row corresponds to an individual athlete competing in an individual Olympic event (athlete-events). The columns are:

ID - Unique number for each athlete
Name - Athlete's name
Sex - M or F
Age - Integer
Height - In centimeters
Weight - In kilograms
Team - Team name
NOC - National Olympic Committee 3-letter code
Games - Year and season
Year - Integer
Season - Summer or Winter
City - Host city
Sport - Sport
Event - Event
Medal - Gold, Silver, Bronze, or NA
'''

#Format
'''
Format dari file tersebut adalah .csv (Comma Separated Values) file
'''
#Sumber
'''
https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results/data#athlete_events.csv
'''

#Dimensi Data ( jumlah baris, jumlah kolom)
print("Dimensi data adalah", df.shape)

#Ukuran file
df.info()
#Dapat diketahui bahwa ukuran file 31.0+ MB


```

```python
#Jelaskan atribut/kolom apa saja yang ada di dalamnya, apa makna tiap atribut mengenai apa, dan jelaskan jenisnya (kategorikal atau kuantitatif, jika kategorikal, apakah nominal, binary, dst.). 

'''
ID - Nomor unik untuk setiap atlet
ID merupakan Data Kuantitatif
Atribut ID terdiri dari

Name - Nama atlet
Name merupakan data kategorikal yaitu nominal

Sex - Jenis Kelamin (Male or Female)
Sex merupakan data kategorikal yaitu nominal

Age - Umur atlet
Age merupakan data kuantitatif

Height - Tinggi atlet (dalam centimeter)
Height merupakan data kuantitatif

Weight - Berat atlet (dalam kilogram)
Weight merupakan data kuantitatif 

Team - Nama Team
Team merupakan data kategorikal yaitu nominal 

NOC - National Olympic Committee 3-letter code
NOC merupakan data kategorikal yaitu nominal

Games - Tahun dan Musim
Games merupakan data kuantitatif dan kategorikal, dimana musim merupakan kategorikal tipe nominal

Year - Tahun
Year merupakan data kuantitatif

Season - Musim ( Summer or Winter)
Season merupakan data kategorikal yaitu nominal

City - Kota tuan rumah
City merupakan data kategorikal  yaitu nominal

Sport -Jenis olahraga
Sport merupakan data kategorikal yaitu nominal

Event - Nama pertandingan
Event merupakan data kategorikal yaitu nominal

Medal - Medali yang didapatkan (Gold, Silver, Bronze, or NA)
Medal merupakan data kategorikal yaitu ordinal"

Penjelasan : 
Data Kuantitatif, adalah data yang berupa angka/numerik
Data kategorik , adalah data yang bukan berupa angka/numerik. 
Data kategorik dibedakan dalam 2 macam data yaitu : 
- Kategorik Nominal, yaitu data kategorik yang tak dapat dinyatakan bahwa kategori yang satu lebih baik dari kategori lainnya"
- Kategorik Ordinal, yaitu data kategorik yang mempunyai urutan tertentu"

'''
minimum,maximum=(df.min(),df.max())

#Range ID
print("Nilai minimum pada ID adalah ", minimum["ID"])
print("Nilai maksimum pada ID adalah ", maximum["ID"])
#Range Age
print("Nilai minimum pada Age adalah ", minimum["Age"])
print("Nilai maksimum pada Age adalah ", maximum["Age"])
#Range Height
print("Nilai minimum pada Height adalah ", minimum["Height"])
print("Nilai maksimum pada Height adalah ", maximum["Height"])
#Range Weight
print("Nilai minimum pada Weight adalah ", minimum["Weight"])
print("Nilai maksimum pada Weight adalah ", maximum["Weight"])
#Range Year
print("Nilai minimum pada Year adalah ", minimum["Year"])
print("Nilai maksimum pada Year adalah ", maximum["Year"])
```
