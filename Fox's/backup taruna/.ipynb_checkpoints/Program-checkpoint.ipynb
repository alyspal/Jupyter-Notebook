```python
#Membagi klasifikasi orang-orang amerika serikat sesuai index
#Agar Tahunnya tetap, tidak ada bbrp row 1 tahun
df2 = df.loc[df["country"] == "United States"]


df21 = df2.loc[(df["age"] == "75+ years") & (df["sex"] == "male")]
df22 = df2.loc[(df["age"] == "55-74 years") & (df["sex"] == "male")]
df23 = df2.loc[(df["age"] == "35-54 years") & (df["sex"] == "male")]
df24 = df2.loc[(df["age"] == "25-34 years") & (df["sex"] == "male")]
df25 = df2.loc[(df["age"] == "15-24 years") & (df["sex"] == "male")]

df26 = df2.loc[(df["age"] == "75+ years") & (df["sex"] == "female")]
df27 = df2.loc[(df["age"] == "55-74 years") & (df["sex"] == "female")]
df28 = df2.loc[(df["age"] == "35-54 years") & (df["sex"] == "female")]
df29 = df2.loc[(df["age"] == "25-34 years") & (df["sex"] == "female")]
df210 = df2.loc[(df["age"] == "15-24 years") & (df["sex"] == "female")]

def call(x):
    if x == 1:
        return df21
    elif x == 2:
        return df22
    elif x == 3:
        return df23
    elif x == 4:
        return df24
    elif x == 5:
        return df25
    elif x == 6:
        return df26
    elif x == 7:
        return df27
    elif x == 8:
        return df28
    elif x == 9:
        return df29
    elif x == 10:
        return df210
    else:
        print("error")
        
print("Index: ")
print("1: 75+ years, male")
print("2: 55-74 years, male")
print("3: 35-54 years,male")
print("4: 25-34 years,male")
print("5: 15-24 years,male")
print()
print("6: 75+ years, female")
print("7: 55-74 years,female")
print("8: 35-54 years,female")
print("9: 25-34 years,female")
print("10: 15-24 years,female")
```

```python
#Testing
print(df21)
```

```python
#Perbandingan kategori suicide_no dan GDP Per Capita
for i in range(1,11):
    x = call(i)
    x.plot(kind="bar",x="gdp_per_capita ($)",y="suicides_no",title="Perbandingan ekonomi dan jumlah kasus bunuh diri "+str(i))

```

```python
#Membandingkan kategori suicide_no dengan populasi
for i in range(1,11):
    x = call(i)
    x.plot(kind="bar",x="population",y="suicides_no",title="Perbandingan ekonomi dan jumlah kasus bunuh diri "+str(i))
```

```python
#Penampilan perubahan suicide_no terhadap waktu
for i in range (1,11):
    x = call(i)
    x.plot(kind="line",x="year",y="suicides_no")
```

```python
#Penampilan perubahan polpulation terhadap waktu
for i in range (1,11):
    x = call(i)
    x.plot(kind="line",x="year",y="population")
```
