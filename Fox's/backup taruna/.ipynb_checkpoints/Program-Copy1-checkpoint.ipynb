```python
import pandas as pd

import os
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("master.csv")

df = df.loc[df.loc[:, 'year'] < 2016,:]

print(df)
```

```python
#Data summary, ada rata-rata, standar devaisi, ekstremum, dan sebagian percentile.
df.describe()
```

```python
#Rata-rata
df.mean()
```

```python
#standar deviasi
df.std()
```

```python
#Percentile (10%, 25%, 50%, 75%, 90%)
df.quantile([.10,.25,.50,.75,.90])
```

```python
#Nilai Minimum
print("Minimum: ")
df.min()
```

```python
#Nilai Maksimum
print("Maximum: ")
df.max()
```

```python
#Perbandingan kategori suicide_no dan age group

sns.barplot(x = 'suicides_no', y = 'age',data = df)
```

```python
#Membandingkan kategori suicide_no dengan jenis kelamin

sns.barplot(x='sex',y = 'suicides_no',data = df)
```

```python
#Penampilan perubahan suicide_no terhadap waktu

sns.lineplot(x='year', y='suicides_no', data=df)
```

```python
#Penampilan perubahan suicides/100k populasi terhadap waktu

sns.lineplot(x='year', y='suicides/100k pop', data=df)
```

```python
#Jumlah data per country
alpha = 0.7
plt.figure(figsize=(10,25))
sns.countplot(y='country', data=df, alpha=alpha)
plt.title('Data by country')
plt.show()
```

```python
#Data genrasi
plt.figure(figsize=(10,25))
df2 = df["generation"].value_counts()
df2.plot(kind = "pie")
plt.title('Data by generations')
plt.show()
```

```python
#Semua korelasi variable yang numerik
plt.figure(figsize=(16,7))
cor = sns.heatmap(df.corr(), annot = True)
```

```python
#Korelasi antara suicide_no dan gdp_per_capita($)
df.plot(kind="scatter",x="gdp_per_capita ($)", y="suicides_no")
```

```python
#Korelasi antara suicide_no/100k populasi dan gdp_per_capita($)
df.plot(kind="scatter",x="gdp_per_capita ($)", y="suicides/100k pop")
```
