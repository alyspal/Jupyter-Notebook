- Pembuatan fungsi agar mempermudah preparing data

```python
def ambil_data_Single (kolom):
    hasil =[0 for i in range(2021)]
    for i in list(kolom):
        temp = data[i]
        for j in range(2021):
            if (temp[j] !=0):
                hasil[j] = temp[0]
    for i in range(2021):
        if hasil[i]==0:
            hasil[i]='Tidak Mengisi'
    return (hasil)
```

```python
def ambil_data_Multi(kolom,kata):
    hasil = [[] for i in range (2021)]
    for i in list(kolom):
        temp = data[i]
        for i in range (2021):
            if (temp[i] !=0):
                hasil[i] = hasil[i] +[kata + str(temp[0])]
            else :
                hasil[i] = hasil[i] +['Tidak '+kata+ str(temp[0])]
    return hasil
```

## Define fungsi Analisis

```python
def CI_solo_rata(xbar,sd,n):
    error = 1.96 * (sd/(n)**(1/2))
    batas_atas = xbar + error
    batas_bawah = xbar - error
    print('Interval Kepercayaan ')
    print('Batas atas = '+str(batas_atas))
    print('Batas Bawah = '+str(batas_bawah))
```

```python
def CI_solo_proporsi(P,n):
    error = 1.96 * (((p)*(1-p))/n)**(1/2)
    batas_atas = p + error
    batas_bawah = p - error
    print('Interval Kepercayaan ')
    print('Batas atas = '+str(batas_atas))
    print('Batas Bawah = '+str(batas_bawah))
```

```python
def CI_ganda_rata(xbar1,xbar2,sd1,sd2,n1,n2):
    sd_gabungan = ((sd1**2)/n1) +((sd2**2)/n2) 
    xbar_gabungan = xbar1-xbar2
    error = 1.96 * (sd_gabungan)
    batas_atas = xbar_gabungan + error
    batas_bawah = xbar_gabungan - error
    print('Interval Kepercayaan ')
    print('Batas atas = '+str(batas_atas))
    print('Batas Bawah = '+str(batas_bawah))
```

```python
def CI_ganda_proporsi(p1,n1,p2,n2):
    p = p1-p2
    p_gabungan = (((p1)*(1-p1))/n1) + (((p2)*(1-p2))/n2)
    error = 1.96*p_gabungan**(1/2)
    batas_atas = p + error
    batas_bawah = p - error
    print('Interval Kepercayaan ')
    print('Batas atas = '+str(batas_atas))
    print('Batas Bawah = '+str(batas_bawah))
```

```python
def uji_statistik(arah,zuji):
    if arah == 'kiri':
        z = 1.65
        if zuji< -z:
            print('H0 ditolak')
        else :
            print('H0 gagal tolak')
    if arah == 'kanan':
        z = 1.65
        if zuji> z:
            print('H0 ditolak')
        else :
            print('H0 gagal tolak')
    if arah== '2way':
        z=1.96
        if ((zuji > (-z)) and (zuji<z) ):
            print('H0 gagal tolak')
        else : 
            print('H0 ditolak')
```

```python
def zuji_1_rata (xbar,miu,sd,n):
    z= ((xbar-miu)*(n**(1/2)))/sd
    return z
```

```python
def zuji_2_rata (xbar1,miu1,sd1,n1,xbar2,miu2,sd2,n2):
    bawah = (((sd1**2)/n1)+((sd2**2)/n2))**(1/2)
    z= (((xbar1-miu1)-(xbar2-miu2))*bawah)
    return z
```

```python
def zuji_1_proporsi (x,n,p):
    pcap = x/n
    bawah = ((pcap)*(1-pcap)/n)**(1/2)
    z = (pcap-p)/bawah
    return z
```

```python
def zuji_2_proporsi(x1,n1,x2,n2):
    pcap1 = x1/n1
    pcap2 = x2/n2
    pcaptotal= (x1+x2)/(n1+n2)
    bawah = ((pcaptotal)*(1-pcaptotal)*((1/n1) + (1/n2)))**(1/2)
    z = (pcap1-pcap2)/bawah
    return z
```
