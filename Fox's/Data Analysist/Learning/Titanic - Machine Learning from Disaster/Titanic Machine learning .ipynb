# feature engineering 

## Nama Keluarga menjadi acuan untuk mengisi NULL pada variabel lain

<b>pemisahan nama keluarga dengan fungsi split  dan menjadikan sebagai new variabel</b>

```python
for i in range (891): # Proses pemisahan nama keluarga
    train.Name[i] = train.Name[i].split(",")
```

```python
Nama_keluarga = [] # variabel yang akan menyimpan nama keluarga

for i in range(len(train.Name)): # Proses pengambilan nama keluarga pada variabel nama
    Nama_keluarga = Nama_keluarga + [train.Name[i][0]]
    
for i in range(len(train.Name)): # Proses Penghapusan nama keluarga pada variabel nama
    train.Name[i] = train.Name[i][1]
```

```python
# Proses pembuatan dataframe Family_Name dan menggabungkan kedalam dataframe train
tempt = pd.DataFrame(Nama_keluarga,columns=["Family_Name"])
train = pd.concat([train,tempt],axis=1)
```

```python
train.head(5)
```

<b> pada data train, nama keluarga pada variabel name telah dihapus dan dibuat menjadi sebuah variabel baru bernama Family_Name </b>

## Pemisahan Title dari variable Name

### Proses Pemisahan menjadi variabel baru

```python
for i in range (891): # Proses pemisahan Title
    train.Name[i] = train.Name[i].split(".")
```

```python
Title = [] # variabel yang akan menyimpan Title

for i in range(len(train.Name)): # Proses pengambilan Title pada variabel nama
    Title = Title + [train.Name[i][0]]
    
for i in range(len(train.Name)): # Proses Penghapusan Title pada variabel nama
    train.Name[i] = train.Name[i][1]
```

```python
# Proses pembuatan dataframe Title dan menggabungkan kedalam dataframe train
tempt = pd.DataFrame(Title,columns=["Title"])
train = pd.concat([train,tempt],axis=1)
```
