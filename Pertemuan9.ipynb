```r
ncs1<-100
xbarcs1<-112
sigmacs1<-100
```

```r
Sirt<-76
Pirt<-180
psirtbar<-Sirt/Pirt
```

```r
SEcs3 = sqrt(psirtbar * (1-psirtbar) / Sirt)   # standard error
SEcs3
```

```r
Ecs3 = qnorm(0.975) * SEcs3       # margin of error
paste("Margin of error", round(Ecs3,4))
```

```r
# estimasi interval proporsi seluruh ibu rumah tangga yang biasa menyaksikan tayangan iklan paling sedikit 2 jam per minggu
paste("Estimasi interval proporsi seluruh ibu rumah tangga yang biasa menyaksikan tayangan iklan paling sedikit 2 jam per minggu berada di antara",
round((psirtbar-E3)*100, 4), " % dan", round((psirtbar+E3)*100, 4), "%")
```

Tanpa nomor 2

```r
alpha<-0.05
pcitizenvote<-0.6
Scitizen<-85
Pcitizen<-148
pcitizenbar<-Scitizen/Pcitizen
```

```r
#H0 (Null Hypothesis) adalah proporsi populasi >= 0.6 atau p>=0.6
zcstn2 = (pcitizenbar-pcitizenvote)/sqrt(pcitizenvote * (1-pcitizenvote)/Pcitizen)
zcstn2   #wilayah kritis
```

```r
z.alphacstn2 = qnorm(1 - alpha) 
Z.lowercstn2 = -z.alphacstn2
paste("Z hitung sebesar",round(zcstn2,4), "dan z alpha 0.05 sebesar", round(Z.lowercstn2,4))
```

Nomor 4

```r
Apunyamobil<-100
APKB<-72
Bpunyamobil<-100
BPKB<-66
```

```r
pA<-APKB/Apunyamobil
pB<-BPKB/Bpunyamobil
```

```r
selisihProporsi = pA - pB  #perbedaan proporsi populasi pertama dan kedua
selisihProporsi
```

```r
batasBawahA = pA - 3 * (sqrt((pA * (1 - pA))/Apunyamobil))
batasAtasA = pA + 3 * (sqrt((pA * (1 - pA))/Apunyamobil))
intervalA = c(batasBawahA, batasAtasA)
intervalA
```

```r
batasBawahB = pB - 3 * (sqrt((pB * (1 - pB))/Bpunyamobil))
batasAtasB = pB + 3 * (sqrt((pB * (1 - pB))/Bpunyamobil))
intervalB = c(batasBawahB, batasAtasB)
intervalB
```

```r
Ecs4 = qnorm(0.95) * sqrt(((pA * (1 - pA))/Apunyamobil) + ((pB * (1 - pB))/Bpunyamobil) ) 
Ecs4
```

```r
batasBawah4 = selisihProporsi-Ecs4
batasAtas4 = selisihProporsi+Ecs4
paste("Beda Proporsi Dua Populasi berada di antara ", round(batasBawah4,4), " dan", round(batasAtas4,4))
```
