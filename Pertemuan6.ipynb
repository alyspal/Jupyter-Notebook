```
install.packages("UsingR")
library(UsingR)
```

```
#Latihan 1
#input data
konsumsi<-read.csv("Lat1.csv")
```

```
head(konsumsi)
```

```
konsumsi<-data.frame(konsumsi)
class(konsumsi)
```

```
library(ggplot2)
dim(konsumsi)
```

```
#define variable
pend<-konsumsi$Pendapatan
peng<-konsumsi$Pengeluaran.untuk.konsumsi
```

```
#Membuat ggplot dari library(ggplot2)
g = ggplot(konsumsi, aes(x=pend, y=peng))
g = g + xlab("Pendapatan (Rp)")
g = g + ylab("Pengeluaran untuk Konsumsi (Rp)")
g = g + geom_point(size=5, colour = "black", alpha=0.5)
g = g + geom_point(size=3, colour="black", alpha=0.2)
g = g + geom_smooth(formula = "y~x", colour="red")
g
```

```
#Model Regresi
fit <- lm(pend~peng, data=konsumsi)
coef(fit)
```

```
#Plot dan garis regresi
plot(pend,peng)
abline(lm(peng~pend))
```

```
#Persamaan Regresi
reg1<-(lm(peng~pend))
reg1
```

```
#Model Regresi secara lengkap
summary(reg1)
```

```
plot(x=pend, y=peng, xlab = "Pendapatan", ylab = "Pengeluaran", col=3, pch=16, main="Korelasi antara Pendapatan dengan Pengeluaran")
cor(pend,peng)
```

```
#Perkiraan besarnya pengeluaran untuk konsumsi jika pendapatannya Rp 100000
Y1<-4.007818+(0.699320*100000)
Y1
```

```
par(mfrow=c(2,2))
plot(reg1)
```
