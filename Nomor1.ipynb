```r
plot(babyfish)
```

```r
kode<-babyfish$Kode
produk<-babyfish$Produk
harga<-babyfish$Harga..Ribu.Rupiah.
penjualanLazana<-babyfish$Volume.Penjualan.di.Lazana
penjualanShoree<-babyfish$Volume.Penjualan.di.Shoree
```

# Scatter Plot

```r
plot(harga,penjualanLazana, xlab = "Harga Produk", ylab = "Volume penjualan di Lazana",
col=3, pch=16, main="Korelasi antara Harga Produk dengan Volume penjualan di Lazana")
abline(lm(harga~penjualanLazana), lwd=2, col="red")
```

```r
plot(harga,penjualanShoree, xlab = "Harga Produk", ylab = "Volume penjualan di Shoree",
col=3, pch=16, main="Korelasi antara Harga Produk dengan Volume penjualan di Shoree")
abline(lm(harga~penjualanShoree), lwd=2, col="red")
```

```r
plot(penjualanLazana,penjualanShoree, xlab = "Volume penjualan di Lazana", ylab = "Volume penjualan di Shoree",
col=3, pch=16, main="Korelasi antara Harga Volume penjualan di Lazana dengan Volume penjualan di Shoree")
abline(lm(penjualanLazana~penjualanShoree), lwd=2, col="red")
```

# Histogram

```r
hist(harga, xlab="Harga", main="Histogram Harga")
lines(density(harga), lwd=2)
```

```r
hist(penjualanLazana, xlab="Volume penjualan di Lazana", main="Histogram Volume penjualan di Lazana")
lines(density(penjualanLazana), lwd=2)
```

```r
hist(penjualanShoree, xlab="Volume penjualan di Shoree", main="Histogram Volume penjualan di Shoree")
lines(density(penjualanShoree), lwd=2)
```
