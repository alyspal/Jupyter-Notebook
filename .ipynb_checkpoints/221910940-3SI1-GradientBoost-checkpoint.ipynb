# Gradient Boosting

## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Sklearn
from sklearn.model_selection import train_test_split      # Membagi data latih dan data uji
from sklearn.model_selection import GridSearchCV          # hyperparameter tuning
from sklearn.metrics import confusion_matrix              # confusion_matrix
from sklearn.metrics import classification_report         # classification_report
```

## Preprosessing

```python
train = pd.read_csv('Train.csv')
train.head()

```

```python
train = train.fillna(train.mode().iloc[0])
```

```python
train.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scale = scaler.fit_transform(train[['Z1','Z2','Z3','Z4']])
```

```python
train_scale= pd.DataFrame(train_scale,columns=['Z1','Z2','Z3','Z4'])
```

```python
train = pd.DataFrame(train,columns=['ID','Provinsi','KodeKabKot','Y','X1','X2','X3','X4','X5','X6'])
```

```python
train = pd.concat([train,train_scale],axis=1)
train
```

```python
bins = [0,14,30,50,80]
labels=[0,1,2,3]
train['X2_Code'] = pd.cut(train['X2'], bins=bins, labels=labels, include_lowest=True)
print (train)
```

```python
X = train.iloc[ : , 3: 15 ]
Y = train.iloc[ : , 3]
```

```python
X=X.drop(axis=1, columns='Y')
```

```python
X=X.drop(axis=1, columns='X2')
```

```python
X
```

```python
Y
```

## Modelling

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 24)

print("X_train shape: ", X_train.shape)
print("Y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("Y_test shape: ", y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
```

```python
gbc.fit(X_train, y_train)
y_pred_gbc = gbc.predict(X_test)
```

```python
from sklearn import metrics
#accuracy
print("acuracy:", metrics.accuracy_score(y_test,y_pred_gbc))
```

```python
print(classification_report(y_test, y_pred_gbc))
```

# Hyperparameter Tunning

```python
GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
parameters = {
    "n_estimators":[5,50,250,500],
    "max_depth":[1,3,5,7,9],
    "learning_rate":[0.01,0.1,1,10,100]
}
```

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(gbc,parameters,cv=5)
cv.fit(X_train,y_train)
```

```python
def display(results):
    print(f'Best parameters are: {results.best_params_}')
    print("\n")
    mean_score = results.cv_results_['mean_test_score']
    std_score = results.cv_results_['std_test_score']
    params = results.cv_results_['params']
    for mean,std,params in zip(mean_score,std_score,params):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
```

```python
display(cv)
```

```python
cv.best_params_
```

```python
# Membuat ulang model GradientBoost

GBC_new = GradientBoostingClassifier(learning_rate=0.1,max_depth=3,n_estimators=500)

GBC_new.fit(X_train, y_train)

y_pred_GBC_new = GBC_new.predict(X_test)
y_train_predict = GBC_new.predict(X_train)
```

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(8,5))
sns.heatmap(confusion_matrix(y_test, y_pred_GBC_new), annot=True, fmt=".0f", ax=ax)
plt.xlabel("y_head")
plt.ylabel("y_true")
plt.show()
```

```python
from sklearn import metrics
#accuracy
print("acuracy:", metrics.accuracy_score(y_test,y_pred_GBC_new))
```

```python
print(classification_report(y_test, y_pred_GBC_new))
```
