## (b) The communalities. 

```r
# Communalities
#get sum of square
getSS <- function(l){return(sum(l^2))}
getCommunality <- function(covMat, L){
    result <- c()
    # number of variables
    p <- ncol(covMat)
    for(i in 1 : p){
        result <- rbind(result, getSS(L[i, ]))
    }
    # rename
    colnames(result) <- c("Communality")
    rownames(result) <- rownames(L)
    return(result)
}
data.frame(getCommunality(corMat, L))
```

## (a) The specific variances. 

```r
getSpecificVariance <- function(covMat, L){
    # communality
    communality <- getCommunality(covMat, L)
    # diagonal elements of covariance matrix : variance of each variable
    variance <- diag(covMat)
    # specific Variance
    specificVar <- variance - communality
    # rename
    colnames(specificVar) <- c("Specific Variance")
    return(specificVar)
}
specificVar <- getSpecificVariance(corMat, L)
data.frame(specificVar)
```

## (c) The proportion of variance explained by each factor. 

```r
getProportionVarianceExplained <- function(covMat, L){
    # total variance
    totalVar <- sum(diag(covMat))
    # proportion of total variance explained by each factor
    f <- ncol(L) # number of factor
    result <- c()
    for(i in 1 : f){
        result <- rbind(result, sum(getSS(L[,i]))/totalVar)
    }
    # rename
    colnames(result) <- c("Proportion of Total Variance Explained")
    rownames(result) <- colnames(L)
    return(result)
}
data.frame(getProportionVarianceExplained(corMat, L))
```

## (d) The residual matrix $R - L_z L_z' - \psi_z$

```r
getResidualMat <- function(covMat, L){
    # number of variable
    p <- ncol(covMat)
    specificVar <- as.vector(getSpecificVariance(covMat, L))
    residualMat <- covMat - L%*%t(L) - diag(specificVar, ncol=p, nrow=p)
    return(residualMat)
}
data.frame(getResidualMat(corMat, L))
```
