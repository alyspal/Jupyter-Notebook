## Bartlett Sphericity Test
Analisis faktor eksplorasi hanya berguna jika matriks korelasi populasi secara statistik berbeda dari matriks identitas. Jika ini sama, variabel-variabelnya sedikit saling terkait, yaitu faktor-faktor spesifik menjelaskan proporsi varians yang lebih besar dan faktor-faktor umum tidak penting. Oleh karena itu, itu harus didefinisikan ketika korelasi antara variabel asli cukup tinggi. Dengan demikian, analisis faktor berguna dalam mengestimasi faktor persekutuan. Dengan pemikiran ini, uji Bartlett Sphericity dapat digunakan. Hipotesisnya adalah:

H0: matriks korelasi populasi sama dengan matriks identitas

H1: matriks korelasi populasi berbeda dengan matriks identitas.

## KMO Test
KMO adalah ukuran kecukupan pengambilan sampel "Kaiser-Meyer-Olkin" dan memeriksa apakah mungkin untuk memfaktorkan variabel utama secara efisien. Matriks korelasi selalu menjadi titik awal. Variabel kurang lebih berkorelasi, tetapi yang lain bisa mempengaruhi korelasi antara dua variabel.Oleh karena itu, dengan KMO, korelasi parsial digunakan untuk mengukur hubungan antara dua variabel dengan menghilangkan pengaruh dari variabel yang tersisa.

```r
#Multicollinearity Test
library("corrplot")

corrplot(cor(sosioeconomics),
 order = "original",
 tl.col = "red",
 tl.srt = 90,
 method="number")
```

Grafik di atas menunjukkan korelasi antara variabel independen (Total Population, Median School Years, Total Employment, Health Services Employment, Median Home Value). Tingginya korelasi antar variabel (Total Population with Median School Years, Total Population with Total Employment, Total Population with Health Service Employment, and Total Employment with Health Services Employment) mungkin menjadi akar penyebab multikolinearitas.

```r
#Homogenity Test
#Homogenity assumption must be fulfilled so the model wouldn't change by every observation (not distracted by time).
uji_bart <- function(x)
{
 method <- "Bartlett's test of sphericity"
 data.name <- deparse(substitute(x))
 x <- subset(x, complete.cases(x))
 n <- nrow(x)
 p <- ncol(x)
 chisq <- (1-n+(2*p+5)/6)*log(det(cor(x)))
 df <- p*(p-1)/2
 p.value <- pchisq(chisq, df, lower.tail=FALSE)
 names(chisq) <- "Khi-squared"
 names(df) <- "df"
 return(structure(list(statistic=chisq, parameter=df, p.value=p.value, method=method, data.name=data.name), class="htest"))
}
uji_bart(sosioeconomics)
```

Dengan tingkat signifikansi 0,05 dan p-value < 0,05, dapat disimpulkan bahwa ada multikolinearitas dalam variabel, dan diperlukan analisis komponen utama.

## 8.14

Perform a principal component analysis using the sample covariance matrix of the sweat data given in Example 5.2. Construct a Q-Q plot for each of the important principal components. Are there any suspect observations? Explain.

```r
sweet_rate <- c(3.7, 5.7, 3.8, 3.2, 3.1, 4.6, 2.4, 7.2, 6.7, 5.4, 3.9, 4.5, 3.5, 4.5, 1.5, 8.5, 4.5, 6.5, 4.1, 5.5)
sodium <- c(48.5, 65.1, 47.2, 53.2, 55.5, 36.1, 24.8, 33.1, 47.4, 54.1, 36.9, 58.8, 27.8, 40.2, 13.5, 56.4, 71.6, 52.8, 44.1, 40.9)
potassium <- c(9.3, 8, 10.9, 12, 9.7, 7.9, 14, 7.6, 8.5, 11.3, 12.7, 12.3, 9.8, 8.4, 10.1, 7.1, 8.2, 10.9, 11.2, 9.4)

sweet_data <- cbind(sweet_rate, sodium, potassium)
sweet_data
```

```r
#Multicollinearity Test
corrplot(cor(sweet_data),
         order = "original",
         tl.col = "red",
         tl.srt = 90,
         method="number")
```

Grafik korelasi di atas menunjukkan korelasi antara variabel independen (Sweet Rate, Sodium, and Potassium). Matriks korelasi menunjukkan bahwa salah satu korelasi berpasangan variabel kurang dari -0,5. Korelasi antara Sweet Rate dan Potassium dapat menunjukkan adanya multikolinearitas pada variabel.

```r
#Homogenity Test
#Homogenity assumption must be fulfilled so the model wouldn't change by every observation (not distracted by time). 
uji_bart <- function(x)
{
  method <- "Bartlett's test of sphericity"
  data.name <- deparse(substitute(x))
  x <- subset(x, complete.cases(x)) 
  n <- nrow(x)
  p <- ncol(x)
  chisq <- (1-n+(2*p+5)/6)*log(det(cor(x)))
  df <- p*(p-1)/2
  p.value <- pchisq(chisq, df, lower.tail=FALSE)
  names(chisq) <- "Khi-squared"
  names(df) <- "df"
  return(structure(list(statistic=chisq, parameter=df, p.value=p.value, method=method, data.name=data.name), class="htest"))
}
uji_bart(sweet_data)
```

Dengan tingkat signifikansi 0,05 dan p-value < 0,05, dapat disimpulkan bahwa ada multikolinearitas dalam variabel, dan perlu dilakukan analisis komponen utama.

```r
#Covariance Matrix
S_sweet <- cov(sweet_data)
S_sweet
```

```r
#Getting the eigen function
sweet_eig <- eigen(S_sweet)
sweet_eig
```

```r
#Eigenvalues 
sweet_eig_val <- sweet_eig$values
sweet_eig_val
```

```r
#Eigenvectors 
sweet_eig_vec <- sweet_eig$vector
sweet_eig_vec
```

```r
prop_var_sweet <- c()
temp <- 0

for(i in c(1:3)){
  temp <- temp + sweet_eig_val[i]
  prop_var <- (temp / sum(sweet_eig_val)) * 100
  prop_var_sweet <- append(prop_var_sweet, prop_var)
}

prop_var_sweet
```

```r
rho_sweet <- matrix(nrow = 3,  ncol = 3)
for(i in c(1:3)){
  for(j in c(1:3))
    rho_sweet[j, i] <- (sweet_eig_vec[j, i] * sqrt(sweet_eig_val[i])) / sqrt(S_sweet[j, j])
}

rho_sweet
```

```r
componets_sweet <- matrix(nrow = 20, ncol = 3)
for(i in c(1:3)){
  for(j in c(1:20)){
    componets_sweet[j, i] <- t(sweet_eig_vec[, i]) %*% sweet_data[j, ]
  }
}
```

```r
pca_sweet_data <- prcomp(sweet_data, scale = F)
fviz_eig(pca_sweet_data)
```

Berdasarkan hasil scree plot dan proporsi variansnya, Komponen utama pertama secara kumulatif menjelaskan 97,17251% dari total varians sampel. Oleh karena itu, dapat disimpulkan bahwa komponen prinsipal yang dipakai untuk menggantikan variabel x adalah komponen prinsipal pertama.
