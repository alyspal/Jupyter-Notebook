## Cleaning

### Standardized

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_datatrainx2 = datatrain[['X2']]
std_datatrainx2
```

```python
scaler = StandardScaler()
std_datatrainx2 = pd.DataFrame(scaler.fit_transform(std_datatrainx2))

std_datatrainx2.columns = std_datatrainx2.columns
std_datatrainx2
```

```python
std_datatrainx2.columns = ['X2']
std_datatrainx2
```

```python
datatrain=datatrain.drop(["X2","Z1","Z2","Z3","Z4",],axis=1)
datatrain=pd.concat([datatrain,std_datatrainx2], axis=1)
datatrain
```

### Creating model dataframe

```python
modeldatatrain=datatrain.drop(["ID","Provinsi"],axis=1)
modeldatatrainx = modeldatatrain.drop(columns='Y')
modeldatatrainy = modeldatatrain['Y']
```

```python
#mdt = modeldatatrain
from sklearn.model_selection import train_test_split
X_trainmdt, X_testmdt, y_trainmdt, y_testmdt = train_test_split(modeldatatrainx, modeldatatrainy, test_size=0.30, random_state=0)
```
