```python
from sklearn import neighbors, datasets, model_selection, metrics
import pandas as pd
```

```python
#Menampilkan Dataset
data=pd.read_csv("/content/Train.csv")
print(data)

datacek = data.copy() #untuk pengecekan data pada preprocessing, sehingga data awal aman
# preview data
datacek.head() #untuk menampilkan 5 data teratas
```

```python
# get some rows informations
print(datacek)

# describe the data
datacek.describe()
```

```python
datacek.info()
```

```python
# data['Y'].value_counts()
for x in data.columns :
  print((data[x] == '?').value_counts())
  print()
```

```python
for x in datacek.columns:
  datacek = datacek[datacek[x] != '?']
```

```python
# get some rows informations
print(datacek)

# describe the data
datacek.describe()
```

```python
datacek.info()
```

```python
datacek.head(10)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for x in data.columns:
  datacek[x] = le.fit_transform(datacek[x])

datacek.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
std_data = pd.DataFrame(scaler.fit_transform(data))

std_data.columns = data.columns
std_data
```

```python
datacekx = datacek.drop(columns='Y')
dataceky = datacek['Y']
datacek
```

```python
datacek
```

```python
import seaborn as sns

sns.set_style('whitegrid')
sns.countplot(x=data['Y'],data=data,palette='RdBu_r')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(datacekx, dataceky, test_size=0.30, random_state=0)
model = neighbors.KNeighborsClassifier(n_neighbors=3)

print(X_train)
print(y_train)

model.fit(X_train, y_train)

predicted = model.predict(X_test)
print(predicted)

# print(metrics.confusion_matrix(y_test, predicted))
# print(metrics.accuracy_score(y_test, predicted))
```

```python
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, predicted, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
```

```python
print(classification_report(y_test,predicted))
```

```python
model.score(X_test, y_test)
```
