## Cleaning

```python
data.head(5)
```

```python
datagraph=data.drop_duplicates()
datagraph.head(5)
```

```python
datagraph = datagraph.drop(['AIRPORT_FLIGHTS_MONTH','AIRLINE_FLIGHTS_MONTH','LATITUDE','LONGITUDE','CARRIER_NAME',
                            'DEPARTING_AIRPORT','PREVIOUS_AIRPORT'], axis=1)
datagraph = datagraph.replace(['EARLY_MORNING', 'MORNING', 'AFTERNOON', 'LATE_NIGHT', 'MIDDAY', 'EVENING'],
                              ['1', '2', '3', '4', '5', '6'])
datagraph.head(5)
```
