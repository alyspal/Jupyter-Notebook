# Preprocessing

```python
sorted(data.DEPARTING_AIRPORT.unique())
```

```python
DA = pd.crosstab(index=data["DEPARTING_AIRPORT"], columns=data["DEP_DEL15"])
DASum = DA[(DA[1]+DA[0])>18000]
for i in DASum[0].keys():
    tempt0 = DASum[0][i]
    tempt1 = DASum[1][i]
    DASum[0][i] =  tempt0
    DASum[1][i] =  tempt1
DASum
```

```python
Atlantadata = data[data['DEPARTING_AIRPORT']=='Atlanta Municipal']
Chicagodata = data[data['DEPARTING_AIRPORT']=="Chicago O'Hare International"]
Dallasdata = data[data['DEPARTING_AIRPORT']=='Dallas Fort Worth Regional']
Douglasdata = data[data['DEPARTING_AIRPORT']=="Douglas Municipal"]
Stapletondata = data[data['DEPARTING_AIRPORT']=="Stapleton International"]
```

```python
busiestdata = pd.concat([Atlantadata, Chicagodata, Dallasdata, Douglasdata, Stapletondata])
busiestdata.head()
```

```python
busiestdata = busiestdata.reset_index()
busiestdata=busiestdata.drop(columns="index")
busiestdata.head()
```

```python
busiestdata.to_csv("busiest_airline_delays.csv",index=False)
```
