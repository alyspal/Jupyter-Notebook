```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import sklearn.preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from matplotlib.colors import ListedColormap
```

```python
#Statistik deskriptif
data.describe()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for x in data.columns:
  data[x] = le.fit_transform(data[x])

data.head(10)
```

```python
#Mengecek missing values
data.isnull().values.any()
```

Karena missing value FALSE, artinya tidak ada missing value yang ditemukan pada dataset.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
std_data = pd.DataFrame(scaler.fit_transform(data))

std_data.columns = data.columns
std_data
```

```python
#Split data menjadi X dan y
X = data.drop(columns='TARGET LABEL (SYMPTOM TIME)')
y = data['TARGET LABEL (SYMPTOM TIME)']
```

```python
#Split X dan y menjadi data training dan data testing
#25% untuk data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
y_train
```

# Training model dengan kernel rbf

```python
#Melakukan training model dengan SVC
model = SVC(kernel = 'rbf', random_state = 0)
model.fit(X_train, y_train)
```

```python
#Memprediksi hasil test data
y_pred =model.predict(X_test)
```

```python
#Membuat confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Training model dengan kernel linear

```python
model2 = SVC(kernel='linear',random_state=0)
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
cm2 = confusion_matrix(y_test, y_pred)
print(cm2)
accuracy_score(y_test,y_pred)
```
