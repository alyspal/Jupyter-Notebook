# AdaBoost and GradientBoost Classifier


# Input packages and datasets

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel('3SI1_Dataset 03_Gejala Ikutan Vaksin.xlsx')
```

# Preprocessing


Checking datasets

```python
df
```

```python
df.describe()
```

Checking null

```python
# data['workclass'].value_counts()
for x in df.columns :
  print((df[x] == '?').value_counts())
  print()
```

**Drop baris dan kolom pada tabel yang mengandung null**

```python
for x in df.columns:
  df = df[df[x] != '?']
```

```python
df = df.dropna()
df = df.drop(columns=['3. Itching'])
df = df.drop(columns=['13. Fainted'])
```

**Lihat tabel dan struktur nya**

```python
# get some rows informations
print(df)

# describe the data
df.describe()
```

**Cek type data**

```python
df.head(10)
```

```python
df.info()
```

Changing categorical data

Data yang dirubah dapat dilihat pada kode dibawah

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for x in df.columns:
  df[x] = le.fit_transform(df[x])

df.head(10)
```

Melihat data setelah diperbaiki

```python
df
```

# Define X and Y

```python
dfx = df.drop(columns='TARGET LABEL (SYMPTOM TIME)')
```

```python
dfy = df['TARGET LABEL (SYMPTOM TIME)']
```

# Split Train and Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfx, dfy, test_size=0.30, random_state=0)
```

# Modeling 

AdaBoost Classifier default parameters


```python
abc = AdaBoostClassifier()
abc.fit(X_train, y_train)
y_pred = abc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
print(classification_report(y_test, y_pred))
```

Parameters tuning with gridsearchcv

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV

model = AdaBoostClassifier(random_state=100)
params = {
    'n_estimators':(50,100),
    'learning_rate':[0.5, 1]
}
gs = GridSearchCV(model, params)
gs.fit(X_train, y_train)
y_pred = gs.predict(X_test)
print('Best parameter: ', gs.best_params_)
```

AdaBoost Classifier with best parameter

```python
abc = AdaBoostClassifier(learning_rate=0.5, n_estimators=50)
abc.fit(X_train, y_train)
y_pred = abc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

GradientBoost with default parameters

```python
gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

Parameters tuning

```python
model = GradientBoostingClassifier(random_state=100)
params = {
        'n_estimators':range(1,50),
        'learning_rate':[0.001, 0.01, 0.1, 0.2, 0.5, 1]
}
gs = GridSearchCV(model, params)
gs.fit(X_train, y_train)
print('Best parameter: ', gs.best_params_)
```

GradientBoost with best parameters


```python
gbc = GradientBoostingClassifier(learning_rate=0.5, n_estimators=38)
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
