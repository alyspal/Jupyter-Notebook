# Perusahaan penghasil drone cerdas


```r
x = c(2100, 2800, 3600, 4000, 4500, 5100, 6800, 7600, 7700, 8500)
y = c(230, 290, 340, 420, 450, 530, 670, 770, 795, 890)
```

```r
plot(x,y)
abline(lm(y~x))
```

```r
#Persamaan Regresi
reg <- lm(y~x)
reg
```

```r
#Model Regresi secara lengkap
summary(reg)
```

```r
plot(x=x, y=y, xlab = "Kapasitas Baterai", ylab = "Jangkauan Terbang",
col=3, pch=16, main="Korelasi antara Kapasitas Baterai dengan Jangkauan Terbang")
cor(x,y)
```
