```r
#Read Data
idi<- read.csv("Indeks Demokrasi Indonesia (IDI) Menurut Aspek dan Provinsi.csv")
```

```r
#Load Data
head(idi)
```

```r
#Transform to 3 dataframe (All, Only Indonesia, All Provinces)
idi<-data.frame(idi)
idinas<-idi[1,]
idiprov<-idi[2:35,]
class(idinas)
class(idiprov)
```

```r
leveneTest(idiprov[2] ~ idiprov[3], data = idiprov) 
```

```r
shapiro.test(idiprov$idiprov[2]) 
```

```r
qqPlot(idiprov$idiprov[2]) 
```

```r
# lm: kode instruksi untuk penyusunan model linier dalam analisa data
model = lm(idiprov[2] ~ idiprov[3], data = idiprov) 
 
# anova: kode instruksi untuk analisa ragam satu jalur
anova(model) 
```

```r
# HSD.test: kode instruksi untuk uji lanjut - Tukey HSD)
(HSD.test(model, "idiprov[3]", unbalanced = FALSE)) 
```
