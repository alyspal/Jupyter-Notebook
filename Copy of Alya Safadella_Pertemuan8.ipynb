# Data

```r
#Read Data
idi<- read.csv("Indeks Demokrasi Indonesia (IDI) Menurut Aspek dan Provinsi.csv")
```

```r
#Load Data
head(idi)
```

```r
#Data Dimension
dim(idi)
```

```r
#Transform to 2 dataframe (All, Only Indonesia, All Provinces)
idi<-data.frame(idi)
idinas<-idi[1,]
idiprov<-idi[2:35,]
class(idinas)
class(idiprov)
```

# Aspek Kebebasan Sipil 2017


```r
#Ukuran Pemusatan 2017
```

```r
#Mean
mean(idiprov$Aspek.Kebebasan.Sipil.2017)
```

```r
#Quantile
quantile(idiprov$Aspek.Kebebasan.Sipil.2017)
```

```r
summary(idiprov$Aspek.Kebebasan.Sipil.2017)
```

```r
#Ukuran Penyebaran
```

```r
#Nilai Minimum dan Provinsinya
min(idiprov$Aspek.Kebebasan.Sipil.2017)
idiprov$Provinsi[which.min(idiprov$Aspek.Kebebasan.Sipil.2017)]
```

```r
#Nilai Maksimum dan Provinsinya
max(idiprov$Aspek.Kebebasan.Sipil.2017)
idiprov$Provinsi[which.max(idiprov$Aspek.Kebebasan.Sipil.2017)]
```

```r
#Rentang
range(idiprov$Aspek.Kebebasan.Sipil.2017)
```

```r
#Standar Deviasi
sd(idiprov$Aspek.Kebebasan.Sipil.2017)
```

```r
#Varians
var(idiprov$Aspek.Kebebasan.Sipil.2017)
```

```r
hist(idiprov[,2], xlab="Aspek Kebebasan Sipil 2017", main="Histogram Aspek Kebebasan Sipil 2017", probability = TRUE, col="grey")
lines(density(idiprov[,2]), lwd=2)
```

```r
idiprov$grpAks2017 <- cut(idiprov$Aspek.Kebebasan.Sipil.2017, breaks = c(0, 60, 80, Inf), labels = c('Buruk', 'Sedang', 'Baik'))
TabgrpAks2017 <- table(idiprov$grpAks2017)
TabgrpAks2017
```

```r
barplot(TabgrpAks2017, xlab = "Aspek Kebebasan Sipil 2017", ylab = "Frekuensi", col=c("red","yellow","green"))
```

```r
pie(TabgrpAks2017, clockwise = T, col=c("red","yellow","green"))
legend("topleft", legend = row.names(TabgrpAks2017), cex=1, title="Aspek Kebebasan Sipil 2017", ncol = 1, bty="n", pch=15, col=c("red","yellow","green"))
```

```r
qqnorm(idiprov$Aspek.Kebebasan.Sipil.2017)
qqline(idiprov$Aspek.Kebebasan.Sipil.2017)
```
